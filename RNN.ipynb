```python
#STRUCTURE OF VANILLA RNN
class RNN:
    #(3,3) is given randomly just for example. When we implement our code, it will be different.
    Whh=np.random.randn(3,3)
    Wxh=np.random.randn(3,3)
    Why=np.random.randn(3,3)
    h=0
    def step(self,x):
        self.h=np.tanh(np.dot(self.Whh,self.h)+np.dot(self.Wxh,x))
        y=np.dot(self.Why,self.h)
        return y
```

```python
r1=RNN()
r2=RNN()
r3=RNN()
r4=RNN()
y=y1=r1.step(np.random.randn(3,1))
y2=r2.step(y1)
y3=r3.step(y2)
y4=r3.step(y3)
```
